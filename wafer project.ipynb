```python
from sklearn.preprocessing import StandardScaler
x=complete_df.loc[:,complete_df.columns].values
x=StandardScaler().fit_transform(x)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
dist=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(normalised_df)
    dist.append(km.inertia_)
plt.plot(range(1,11),dist,marker='o')
plt.show()
```

```python
kn=KMeans(n_clusters=5)
y_pred=kn.fit_predict(normalised_df)
y_pred
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler


x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=2,test_size=0.2)
```

```python
minmax=MinMaxScaler()
x_train_scaled=minmax.fit_transform(x_train)
x_test_scaled=minmax.fit_transform(x_test)
x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
lg=LogisticRegression()
lg.fit(x_train_scaled,y_train)

dt=DecisionTreeClassifier()
dt.fit(x_train_scaled,y_train)

rf=RandomForestClassifier()
rf.fit(x_train_scaled,y_train)

gd=GradientBoostingClassifier()
gd.fit(x_train_scaled,y_train)

svc=SVC()
svc.fit(x_train_scaled,y_train)

kn=KNeighborsClassifier()
kn.fit(x_train_scaled,y_train)
```

```python
print('logisticReg score',lg.score(x_test_scaled,y_test))
print('decisionTree score',dt.score(x_test_scaled,y_test))
print('randomforest score',rf.score(x_test_scaled,y_test))
print('gradientboosting score',gd.score(x_test_scaled,y_test))
print('svc score',svc.score(x_test_scaled,y_test))
print('KNeighbors score',kn.score(x_test_scaled,y_test))
```
